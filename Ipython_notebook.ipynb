```python
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
from matplotlib import style

data = {'Alopathic':250, 'Ayurvedic':200, 'Unani':100, 'Homeo':150}
category = list(data.keys())
doctors_count = list(data.values())

style.use('ggplot')
plt.figure(figsize=(5,5))
color = ['royalblue', 'red', 'purple', 'yellow']
plt.bar(category, doctors_count, color=color, width=0.5)
plt.xlabel('Category')
plt.ylabel('Doctors Count')
plt.title('Doctors count of each category of treatments')
```
